```python
experience_prompt = f"""
You are an AI system that extracts professional experience from profiles.

TASK:
Analyze the user profile and estimate experience.

USER PROFILE:
{user_profile}

INSTRUCTIONS:
1. Estimate total years of experience.
2. Classify seniority level.
3. Provide experience breakdown by area.

Respond ONLY in JSON:

{{
  "total_experience_years": <number>,
  "seniority_level": "<Junior / Mid / Senior>",
  "experience_breakdown": {{
    "software_development": "<Low/Medium/High>",
    "cloud": "<Low/Medium/High>",
    "leadership_or_mentoring": "<Low/Medium/High>"
  }}
}}
"""


competency_prompt = f"""
You are an AI system that identifies core competencies from profiles.

TASK:
Extract the main technical and professional competencies.

USER PROFILE:
{user_profile}

INSTRUCTIONS:
1. Identify core competencies.
2. Identify secondary skills.
3. Assess confidence level.

Respond ONLY in JSON:

{{
  "core_competencies": [
    "Competency 1",
    "Competency 2"
  ],
  "secondary_skills": [
    "Skill 1",
    "Skill 2"
  ],
  "confidence": "<High / Medium / Low>"
}}
"""
role_prompt = f"""
You are an AI career advisor.

TASK:
Recommend suitable job roles based on the user profile.

USER PROFILE:
{user_profile}

INSTRUCTIONS:
1. Suggest 3 suitable roles.
2. Explain why each role fits.
3. Indicate readiness level.

Respond ONLY in JSON:

{{
  "recommended_roles": [
    {{
      "role": "Role Name",
      "fit_reason": "Short explanation",
      "readiness": "<Ready / Almost Ready / Needs Upskilling>"
    }}
  ]
}}
"""
```

```python
from openai import OpenAI

#client = OpenAI()
client = OpenAI( base_url = os.environ["OPENAI_API_BASE"], api_key = os.environ["OPENAI_API_KEY"])

def run_prompt(prompt):
    response = client.responses.create(
        model="glm4:latest",
        input=prompt,
        max_output_tokens=300
    )
    return response.output_text


```

```python
experience_output = run_prompt(experience_prompt)
print("=== EXPERIENCE ===")
print(experience_output)
```

```python
competency_output = run_prompt(competency_prompt)
print("\n=== CORE COMPETENCIES ===")
print(competency_output)
```

```python
role_output = run_prompt(role_prompt)
print("\n=== RECOMMENDED ROLES ===")
print(role_output)
```
